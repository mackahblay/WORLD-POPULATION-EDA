analyzing 2024 population

```python
plt.figure(figsize = (15, 6))
sns.histplot(df['2023 population'], bins = 15, edgecolor = 'black', color = 'navy', kde = True)
plt.title('2023 population distribution')
plt.show()
```

looking at the **total** 2023 populations by continent

```python
continent_total = df.groupby('continent')['2023 population'].sum().sort_values(ascending = False)
continent_total
```

plotting with pie chart

```python
plt.pie(continent_total, labels = continent_total.index, autopct = '%1.1f%%')
plt.title('Populations by Continents - Visualized')
plt.show()
```

```python
plt.figure(figsize = (15, 6))
sns.lineplot( x = df['continent'], y = df['2023 population'], label = '2023')
sns.lineplot( x = df['continent'], y = df['2022 population'], label = '2022')
sns.lineplot( x = df['continent'], y = df['2020 population'], label = '2020')
sns.lineplot( x = df['continent'], y = df['2015 population'], label = '2015')
plt.title('2015-2023 populations by continent\n\n')
plt.xlabel('\n\n\ncontinent')
plt.ylabel('population')
plt.xticks(rotation = 25)
plt.legend(loc = 'upper right')
plt.show()
```

top 10 most **populous** countries in 2023

```python
top_10_2023 = df.sort_values(by = '2023 population', ascending = False).head(10).set_index('country')
top_10_2023
```

visualizing most populous countries

```python
cols = ['pink', 'navy', 'orange', 'lime', 'red', 'gray', 'purple', 'black']
top_10_2023[ '2023 population'].plot(kind = 'barh', color = cols)
plt.title('\ntop 10 most populous countries in 2023\n')
plt.xlabel('\npopulation (in billions)\n')
plt.show()
```

analyzing continents by size

```python
df.groupby('continent')['area (km²)'].sum().sort_values(ascending = False)
```

analyzing continents by density

```python
df.groupby('continent')['density (km²)'].sum().sort_values(ascending = False)
```

**visualizing continents by area**

```python
plt.figure(figsize = (13, 4))
sns.violinplot(x = 'continent', y = 'area (km²)', data = df, palette = 'Set2', hue = 'continent')
plt.title('\nviolin plot of continents and area in km\n')
plt.xticks(rotation = 25)
plt.xlabel('\ncontinent')
plt.ylabel('area size (km²)')
plt.show()
```

**visualizing continents by area**

```python
plt.figure(figsize = (13, 4))
sns.boxplot(x = 'continent', y = 'density (km²)', data = df, palette = 'Set2', hue = 'continent', showfliers = False)
plt.title('\nboxplot plot of continents and density in km²')
plt.xticks(rotation = 25)
plt.xlabel('\ncontinent')
plt.ylabel('density in (km²)')
plt.show()
```
